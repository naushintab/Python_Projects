```python
### Part 02 : Cleaning data + Generating Heat map
```

```python
df
```

```python
df.head()
```

```python
heat_df = df[df['ap_lo'] <= df['ap_hi']]
heat_df
```

```python
heat1_df = heat_df[heat_df['height'] >= heat_df['height'].quantile(0.025)] 
heat1_df
```

```python
heat2_df = heat1_df[heat_df['height'] <= heat1_df['height'].quantile(0.975)]
heat2_df
```

```python
heat3_df = heat2_df[heat2_df['weight'] >= heat2_df['weight'].quantile(0.025)]
heat3_df
```

```python
heat4_df = heat3_df[heat3_df['weight'] <= heat3_df['weight'].quantile(0.975)]
heat4_df
```

```python
type(heat4_df)
```

```python
heat4_df.info()
```

```python
corr_matrix = heat4_df.corr(method='pearson', min_periods=1)
corr_matrix
```

```python
# corr = np.corrcoef(np.random.randn(10, 200))
corr = round(corr_matrix,5)
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style('ticks'):
    f, ax = plt.subplots(figsize=(7, 7))
    ax = sns.heatmap(corr, mask=mask, vmin=-0.1, 
                     vmax=.3, annot=True, cbar_kws={'shrink':.40},
                     square=True, center=0, fmt='.1f', linewidths='0.4',
                     robust=False)
```
